## 1. Logistic Map: time series when $r$ = 3.5

The logistic map equation is: $f(x)=rx(1-x)$

The $r$ value sets the parameter. By picking a value and keeping it constant one can observe what effect it has on an initial condition or seed's $x_0$ orbit over time.

With an initial condition of $x_0$ = 0.47 after a 100 iterations, one can see that the orbit is pulled toward a period cycle of four points approximating 0.5, 0.9, 0.4, and 0.8 in that order. This result is consistent no matter the initial condition chosen, or how many iterations are conducted, as long as the $r$ value is kept the same.

```python
# time series
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors

# first define a function / dyanmical system
def f(x):
    return r * x * (1-x)

r =  3.5
start = 0
end = 100
seed = 0.47
NumIteration = int(100)

# initialization
x_values = np.zeros(NumIteration + 1)
y_values = np.zeros(NumIteration + 1)
x_values[0] = seed


# iteration
for i in range(NumIteration):
    y_values[i] = f(x_values[i])
    x_values[i+1] = y_values[i]


plt.figure(figsize=(16,3.5))
plt.plot(np.arange(start,end), x_values[start:NumIteration], 'H', mfc = 'tab:pink', mec = 'tab:pink', markersize=7)
plt.plot(np.arange(start,end), x_values[start:NumIteration], '--', color = 'tab:purple', lw=2)
plt.title(f"Logistic Time Series, with $r$ = {r} and $x_0$ = {seed}", fontsize=17)
plt.xlabel("Time $t$", fontsize=15)
plt.ylabel("$X_{t}$", fontsize=15)

plt.show()
y1 = y_values
print(y_values)
```

## Logistic Map: time series when $r$ = 4.0

With the same initial condition $x_0$ = 0.47, but increasing the $r$ parameter to 4.0, the orbit tends toward chaos. This result is consistent no matter the initial condition chosen or how many iterations are conducted.

Any $r$ value above 4.0 will tend toward negative infinity and is not of interest for this study.

```python
# time series
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors

# first define a function / dyanmical system
def f(x):
    return r * x * (1-x)

r =  4.0
start = 0
end = 100
seed = 0.47
NumIteration = int(100)

# initialization
x_values = np.zeros(NumIteration + 1)
y_values = np.zeros(NumIteration + 1)
x_values[0] = seed


# iteration
for i in range(NumIteration):
    y_values[i] = f(x_values[i])
    x_values[i+1] = y_values[i]


plt.figure(figsize=(16,3.5))
plt.plot(np.arange(start,end), x_values[start:NumIteration], 'H', mfc = 'tab:pink', mec = 'tab:pink', markersize=7)
plt.plot(np.arange(start,end), x_values[start:NumIteration], '--', color = 'tab:purple', lw=2)
plt.title(f"Logistic Time Series, with $r$ = {r} and $x_0$ = {seed}", fontsize=17)
plt.xlabel("Time $t$", fontsize=15)
plt.ylabel("$X_{t}$", fontsize=15)

plt.show()
y1 = y_values
```

##Bifurcation Diagram

When taking the final states of orbits of multiple initial conditions between $0$ and $1$ ($y$ axis) against the different $r$ values between $0$ and $4$ ($x$ axis), one gets a visualization of the logistic map also known as a bifurcation diagram.

```python
import numpy as np
import matplotlib.pyplot as plt

interval = (0, 4)  # start, end
accuracy = 0.0001
reps = 600  # number of repetitions
numtoplot = 500
lims = np.zeros(reps)

fig, biax = plt.subplots()
fig.set_size_inches(16, 9)

lims[0] = np.random.rand()
for r in np.arange(interval[0], interval[1], accuracy):
    for i in range(reps - 1):
        lims[i + 1] = r * lims[i] * (1 - lims[i])

    biax.plot([r] * numtoplot, lims[reps - numtoplot :], "b.", markersize=0.02)

biax.set_xlabel("r", fontsize=18)  # Adjust the fontsize here
biax.set_ylabel("x", fontsize=18)  # Adjust the fontsize here
biax.set_title("Logistic Map", fontsize=24)  # Adjust the fontsize here
plt.show()
```

##Zoom in for $r$ values 2.8 to 4 and Sharkovsky Ordering

The periodic cycles have been observed to follow reverse Sharkovsky ordering.

There is a way of ordering integers known as Sharkovsky ordering, which is as follows:

3, 5, 7, 9, ... (odd numbers)

2×3, 2×5, 2×7, 2×9,... (2 times odd numbers)

4×3, 4×5, 4×7, 4×9,... (4 times odd numbers)

8×3, 8×5, 8×7, 8×9,... (8 times odd numbers)

.

.

.

... $2^4, 2^3, 2^2, 2, 1$,... (decreasing powers of 2) .

The first period we see as we increase $r$ is 1. This doubles to 2, then 4, 8, 16, and so on. The last number in reverse Sharkovsky ordering is 3. Accordingly, moving left to right, the period 3 window is the last new window to appear.

```python
import numpy as np
import matplotlib.pyplot as plt

interval = (2.8, 4)  # start, end
accuracy = 0.0001
reps = 600  # number of repetitions
numtoplot = 500
lims = np.zeros(reps)

fig, biax = plt.subplots()
fig.set_size_inches(16, 9)

lims[0] = np.random.rand()
for r in np.arange(interval[0], interval[1], accuracy):
    for i in range(reps - 1):
        lims[i + 1] = r * lims[i] * (1 - lims[i])

    biax.plot([r] * numtoplot, lims[reps - numtoplot :], "b.", markersize=0.02)

biax.set_xlabel("r", fontsize=18)  # Adjust the fontsize here
biax.set_ylabel("x", fontsize=18)  # Adjust the fontsize here
biax.set_title("Logistic Map", fontsize=24)  # Adjust the fontsize here
plt.show()
```

## Histograms of multiple orbits similar to flipping a coin and reveal connection to PRNGs

Even with different initial conditions $x_0$ (0.2, 0.4, 0.8, and 0.9) iterated a million times, each histogram of the logistic map at $r = 4$ reveals the same frequency on the $x$ interval from 0 to 1. If cut in half the histogram even looks symmetrical on each side. To check if this is indeed the case the histogram of initial condition $x_0 =0.2$ is chosen and the frequency the orbit spends below and above 0.5 is checked, revealing the two bins are practically identical with a density of 50.027100000000004%

The chances of getting a number above or below 0.5 are equivalent to flipping a coin where one would have a 50% chance of getting heads or tails. This is profound as the logistic map is not created by chance at all. It comes from a deterministic equation that follows a predictable pattern.

And this is how Pseudo-Random Number Generators (PRNGs) are created! This is the reason that for ANY pseudo-random number generator, in any programming language, you can assign any seed. Different seeds will give you random numbers, but using the same seed will always give you the same random sequence.

```python
import numpy as np
import matplotlib.pyplot as plt
import timeit


# first define a function / dynamical system
def f(x):
    return r * x * (1 - x)

# define the function that compute y_values for different seeds
def garden(seed, NumIteration):

    start = timeit.default_timer()


    # initialization
    x_values = np.zeros(NumIteration + 1)
    y_values = np.zeros(NumIteration + 1)
    x_values[0] = seed

    # iteration
    for i in range(NumIteration):
        y_values[i] = f(x_values[i])
        x_values[i+1] = y_values[i]

    stop = timeit.default_timer()


    return x_values[0:-1], stop - start  # return y_value and computing time


def pseudo(y_values):

    # find the length of the y_values
    length = np.shape(y_values)[0]

    #initialize a vector of 0s to be returned
    vector_01 = np.zeros(length,  dtype=int)

    for i in range(length):
        # if y_values[i] > 0.5:  # this if for r = 4.0
        if y_values[i] > 0.7:  # this if for r = 3.625
            vector_01[i] = 1

    total_1s = np.sum(vector_01)
    density_1 = total_1s / length
    print(vector_01[length - 50 : length])
    print(f"The density of 1 is: {100*density_1}%.")

    return vector_01



def pseudo2(y_values):

    # find the length of the y_values
    length = np.shape(y_values)[0]

    #initialize a vector of 0s to be returned
    vector_01 = np.zeros(length,  dtype=int)

    for i in range(length):
        b = ((y_values[i] * 100 ) // 1) % 10
        if b % 2 == 1:
            vector_01[i] = 1

    total_1s = np.sum(vector_01)
    density_1 = total_1s / length
    print(vector_01[length - 20 : length])
    print(f"The density of 1 is: {100*density_1}%./n")

    return vector_01

# global parameters
# r = 3.5  # period-4
# r = 4.0  # chaotic, with invariant distribution -> pseudo random number generator
r = 4
NumIteration = int(1E6)



num_plots = 4
# seed_list = [0.8, 0.8001, 0.800001, 0.80000001]
# seed_list = [0.3, 0.300001, 0.300000001, 0.5]
seed_list = [0.2, 0.4, 0.8, 0.9]
# plt.figure(figsize=(160, 22.5 * num_plots))
plt.figure(figsize=(16, 2.25 * num_plots))


random_vector_array = np.zeros( (num_plots, NumIteration) )

for i in range(num_plots):
    seed = seed_list[i]

    plt.subplot(num_plots, 2, 2*i + 1)
    plot_step_size = 10001

    y_values, computing_time = garden(seed, NumIteration)
    print(y_values[0:20])
    plt.plot(y_values[0 : NumIteration : plot_step_size], linewidth=2.0)   # skip the last entry

    title_str1 = f"Logistic Time series\n"
    title_str2 = f"with r = {r}, seed = {seed}"
    plt.title(title_str1 + title_str2)
    plt.xlabel('Iteration')
    plt.ylabel('Value')

    # plot histogram
    plt.subplot(num_plots, 2, 2*i + 2)

    # bins divided evenly
    plt.hist(y_values[0 : NumIteration], range=[0, 1], bins = 200)

    # for r=3.625 with two disjoint intervals of end points
    # plt.hist(y_values[0 : NumIteration], range=[0, 1], bins = [0, 0.69, 0.71, 1], edgecolor='black')

    title_str1 = (f"Histogram of Logistic Map\n")
    title_str2 = (f"with {NumIteration} Iterations")
    plt.title(title_str1 + title_str2)
    plt.xlabel('Value')
    plt.ylabel('Frequency')


    # generate pseudo-random numbers
    a = pseudo2(y_values)
    random_vector_array[i, :] = a
    # plt.plot(random_vector, ".")



# plot time series
plt.tight_layout()
plt.show()
```

### Insensitivity to Initial Conditions: When $r = 3.5$, the numbers generated are not random

```python
def logistic_map(seed, r=3.5, n=1000000):
    x = seed
    prns = []
    for _ in range(n):
        x = r * x * (1 - x)
        prns.append(int(x > 0.5))
    density_of_1 = sum(prns) / n * 100
    return prns, density_of_1

def print_results(seed):
    prns, density_of_1 = logistic_map(seed)
    print(f"Seed = {seed}:")
    print(f"The total PRN generated are {len(prns)}; the last 20 pseudo random numbers are:")
    print(f"{prns[-20:]}")
    print(f"The density of 1 is: {density_of_1:.4f}%.\n")

seeds = [0.8, 0.8001, 0.800001, 0.8000001]

for seed in seeds:
    print_results(seed)
```
